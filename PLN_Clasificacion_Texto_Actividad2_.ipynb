# Redes Kolmogorov-Arnold (KAN)

## Arquitectura
Las **Redes Kolmogorov-Arnold (KAN)** utilizan funciones de activación aprendibles en los "pesos", en contraste con las perceptrones multicapa (MLP) que utilizan funciones de activación fijas en los nodos. Cada parámetro de peso en una KAN se reemplaza por una función univariante parametrizada como spline, lo que proporciona una flexibilidad notable para la adaptación local sin afectar otras regiones de la red.

## Beneficios
1. **Interpretación Visual**: Las funciones spline permiten una interpretación visual clara de las transformaciones de entrada, facilitando modificaciones y comprensión detallada del proceso de transformación de datos.
2. **Leyes de Escalado Neuronal**: Las KAN tienen leyes de escalado más favorables que las MLP, manteniendo o mejorando el rendimiento efectivamente al escalar la red.
3. **Eficiencia en el Aprendizaje**: Requieren menos épocas para converger a soluciones óptimas y presentan mayor estabilidad durante el entrenamiento.

## Limitaciones
1. **Complejidad en la Configuración**: La optimización y parametrización de funciones univariantes pueden añadir una capa de complejidad al diseño y entrenamiento.
2. **Riesgo de Sobreajuste**: Las funciones spline permiten ajustes locales intensivos, que pueden llevar a sobreajuste si no se manejan adecuadamente.

## Casos de Uso
1. **Descubrimiento Científico**: En campos como la física y la química, las KAN ayudan a descubrir leyes físicas y propiedades químicas a partir de datos experimentales.
2. **Aplicaciones de Aprendizaje Automático**: Las KAN son efectivas en aprendizaje automático tradicional como reconocimiento de imágenes y voz, aprendiendo representaciones más efectivas con menos iteraciones de entrenamiento.
3. **Optimización en Ingeniería y Física**: Aplicadas en la simulación de dinámica de fluidos y optimización estructural, las KAN logran mayor precisión y requieren menos recursos computacionales que modelos MLP equivalentes.

# Comparación entre MLP y KAN

## Teoría y Fórmula

- **MLP (Perceptrón Multicapa)**:
  - Basado en el **Teorema de Aproximación Universal**.
  - Fórmula:$$ f(x) = \sum_{i=1}^{N_c} \sigma_i(w_i \cdot x + b_i) $$ donde **$$\sigma$$** son funciones de activación fijas y  **$$w_i$$ y $$b_i$$** son pesos y sesgos aprendibles.

- **KAN (Red Kolmogorov-Arnold)**:
  - Basado en el **Teorema de Representación de Kolmogorov-Arnold**.
  - Fórmula: $$ f(x) = \sum_{q=1}^{2n+1} \phi_q(\sum_{p=1}^n a_{qp}x_p) $$
  - **$$\phi$$** son funciones de activación aprendibles en los "pesos".
  - La suma se realiza en los nodos.

## Modelos Visualizados

- **MLP**:
  - **Modelo Somero**: Una sola capa oculta con funciones de activación fijas y pesos lineales aprendibles.
  - **Modelo Profundo**: Arquitectura multicapa con múltiples capas ocultas, cada una con pesos lineales y funciones de activación no lineales fijas.

- **KAN**:
  - **Modelo Somero**: Una única capa con activaciones no lineales aprendibles.
  - **Modelo Profundo**: Múltiples capas con funciones de activación aprendibles en cada conexión, ofreciendo mayor flexibilidad.

## Diferencias Clave

- **MLP**:
  - Las funciones de activación son fijas y no se adaptan durante el entrenamiento.
  - El aprendizaje ocurre principalmente en los pesos lineales entre las capas.

- **KAN**:
  - No utilizan pesos lineales; cada conexión entre nodos tiene una función de activación que se optimiza durante el entrenamiento.
  - Ofrece una estructura más flexible y potencialmente más poderosa para aprender representaciones de datos.

Esta comparación ilustra cómo las **Redes Kolmogorov-Arnold (KAN)** pueden proporcionar una estructura más flexible y potencialmente más eficaz para modelar relaciones complejas y no lineales en los datos, en comparación con los tradicionales **Perceptrones Multicapa (MLP)**.
