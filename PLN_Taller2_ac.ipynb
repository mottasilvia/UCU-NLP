```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re

# Inicializar una lista para almacenar las reseñas
all_reviews = []

# Función para extraer reseñas de una página de IMDb
def extract_reviews(movie, url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')


 # Encontrar todas las reseñas en la página
    reviews = soup.find_all('div', class_='text show-more__control')
    for review in reviews:
        all_reviews.append({
            'movie': movie,
            'movie_url': url,
            'review': review.get_text()
        })

# Extraer reseñas para cada película en la lista
for movie in movie_urls:
    extract_reviews(movie, movie_urls[movie])

# Convertir la lista de reseñas en un DataFrame de pandas
df_reviews = pd.DataFrame(all_reviews)

# Guardar las reseñas en un archivo CSV
df_reviews.to_csv('reseñas_películas_imdb.csv', index=False)

print(f"Recolectadas {len(all_reviews)} reseñas.")

df_reviews.info()
print(df_reviews.head())
print(df_reviews.tail())
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

# Descargar los recursos necesarios de nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Inicializar el lematizador y obtener la lista de stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Función para limpiar texto
def clean_text(text):
    # Convertir a minúsculas
    text = text.lower()
    # Eliminar caracteres especiales, números y guiones bajos
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\d', ' ', text)
    text = re.sub(r'_', ' ', text)
    # Eliminar puntuación
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Función para eliminar stop words y lematizar
def preprocess_text(text):
    # Tokenizar el texto
    words = nltk.word_tokenize(text)
    # Eliminar stop words y lematizar
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)

# leer csv
df_reviews = pd.read_csv('reseñas_películas_imdb.csv')

clean_reviews = []
all_words = []
words_without_sw = []
lemmas = []

for review in df_reviews["review"]:
  clean_review = clean_text(review)

  all_words_review = nltk.word_tokenize(clean_review)

  words_without_sw_review = [word for word in all_words_review if word not in stop_words]

  lemmas_review = [lemmatizer.lemmatize(word) for word in words_without_sw_review]

  clean_reviews.append(clean_review)
  all_words.append(" ".join(all_words_review))
  words_without_sw.append(" ".join(words_without_sw_review))
  lemmas.append(" ".join(lemmas_review))

df_reviews['clean_review'] = clean_reviews
df_reviews['all_words'] = all_words
df_reviews['words_without_sw'] = words_without_sw
df_reviews['lemmas'] = lemmas

# Guardar las reseñas preprocesadas en un archivo CSV
df_reviews.to_csv('reseñas_películas_imdb_preprocesadas.csv', index=False)

df_reviews.info()
print(df_reviews.head())
print(df_reviews.tail())

# TODO steam
```

```python
# Crear la matriz de términos de documentos utilizando BoW
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df_reviews['lemmas'])

# Convertir la matriz a un DataFrame para visualizar mejor
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Guardar la matriz BoW en un archivo CSV
bow_df.to_csv('bow_reseñas_películas_imdb.csv', index=False)

print("Matriz de términos de documentos (BoW) creada y guardada en 'bow_reseñas_películas_imdb.csv'.")

print(bow_df.info())
print(bow_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sumar las ocurrencias de cada palabra
word_counts = bow_df.sum().sort_values(ascending=False)

# Mostrar las 20 palabras más frecuentes
top_words = word_counts.head(20)

# Crear un gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=top_words.values, y=top_words.index, palette="viridis", hue=top_words.index)
plt.title('Top 20 palabras más frecuentes en las reseñas')
plt.xlabel('Frecuencia')
plt.ylabel('Palabras')
plt.show()
```

```python
import numpy as np
import gensim.downloader as api


# Descargar el modelo preentrenado de GloVe
glove_model = api.load("glove-wiki-gigaword-100")


# Función para obtener el embedding de una palabra
def get_word_embedding(word):
    try:
        return glove_model[word]
    except KeyError:
        return np.zeros(glove_model.vector_size)

# Función para obtener el embedding de una reseña
def get_review_embedding(review):
    embeddings = [get_word_embedding(word) for word in review]
    if embeddings:
        return np.mean(embeddings, axis=0)
    else:
        return np.zeros(glove_model.vector_size)

# Obtener los embeddings para todas las reseñas
df_reviews['embedding'] = df_reviews['lemmas'].apply(get_review_embedding)

# Guardar los embeddings en un archivo CSV
embeddings_df = pd.DataFrame(df_reviews['embedding'].tolist())
embeddings_df.to_csv('embeddings_reseñas_películas_imdb.csv', index=False)

print("Embeddings generados y guardados en 'embeddings_reseñas_películas_imdb.csv'.")

print(embeddings_df.info())
print(embeddings_df.head())
```

```python
from sklearn.decomposition import PCA

# Asegurarse de que los embeddings están en formato numpy array
embeddings = embeddings_df.values

# Aplicar PCA para reducir la dimensionalidad a 2 componentes
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings)

# Crear un DataFrame con los resultados de PCA
pca_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])

# Visualizar los resultados de PCA en 2D
plt.figure(figsize=(12, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], alpha=0.5)
plt.title('Visualización de Embeddings con PCA (2D)')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
from sklearn.manifold import TSNE

# Aplicar t-SNE para reducir la dimensionalidad a 2 componentes
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(embeddings)

# Crear un DataFrame con los resultados de t-SNE
tsne_df = pd.DataFrame(tsne_result, columns=['t-SNE1', 't-SNE2'])

# Visualizar los resultados de t-SNE en 2D
plt.figure(figsize=(12, 6))
plt.scatter(tsne_df['t-SNE1'], tsne_df['t-SNE2'], alpha=0.5)
plt.title('Visualización de Embeddings con t-SNE (2D)')
plt.xlabel('t-SNE1')
plt.ylabel('t-SNE2')
plt.show()
```
